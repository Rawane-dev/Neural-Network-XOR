```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])
model=Sequential([
    Dense(4,activation='tanh',input_shape=(2,)),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X, y,

          epochs=1000, 
          verbose=0,
          
)
```

```python
loss, accuracy = model.evaluate(X, y)
```

```python
pred=model.predict(X)
print(f"  Final Loss     : {loss:.6f}")
print(f"  Final Accuracy : {accuracy * 100:.1f}%")
```

# XOR Neural Network

## Forward Propagation
During forward propagation, input data (XOR pairs [0,0], [0,1], [1,0], [1,1]) flows through the network's layers. The inputs first reach the hidden layer with 4 neurons using tanh activation, which transforms the data through weighted connections. These transformed values then pass to the output layer with a single neuron using sigmoid activation, producing predictions between 0 and 1 for each input pattern.

## Error Computation
The network computes error by comparing its predictions against the true XOR outputs ([0,1,1,0]) using binary cross-entropy loss. This loss function quantifies the difference between predicted and actual values, with higher values indicating worse performance. As shown in the training output, the loss starts at 0.7643 and gradually decreases to 0.5500 by epoch 1000.

## Backpropagation
In backpropagation, the network calculates how each weight contributed to the error by applying the chain rule backwards through the network layers. It determines the gradient of the loss with respect to every weight, identifying which weights need adjustment and in which direction to reduce the overall error.

## Weight Updates
The Adam optimizer updates the network weights after each epoch using the calculated gradients. Weights are adjusted in small steps to minimize the error, which explains the steady loss decrease from 0.7643 to 0.5500 over 1000 epochs. Despite these updates, the network only achieved 75% accuracy, successfully learning 3 of the 4 XOR patterns.
